```python
import numpy as np
from math import pi
import pandas as pd
import matplotlib.pyplot as  plt

df = pd.read_csv("datos.csv")
```

```python
def qi(a, V):
    ro = 817
    sigma = 1.29
    g = 9.81
    d = 5e-3
    return (4*pi*(a**3)*(ro-sigma)*g*d)/(3*V)

def qi_star(q, e):
    return int(q/e + 0.5)
```

```python
e = np.arange(1e-19, 1.7e-19, 0.001e-19)
```

```python
data = df[['Tension de Equilibrio (V)','radio(m)']]
a = list(data['radio(m)'])
print(a)
v = list(data['Tension de Equilibrio (V)'])
print(v)
```

```python
fe = []
approx = []
for i in e:
    sum = 0
    for j in range(0, len(a)):
        sum += (qi(a[j], v[j])/i - qi_star(qi(a[j], v[j]), i))**2
        j += 1
    i+=1
    fe.append(sum)

for i in fe:
    approx.append(abs(i-1.602))
    i+=1

b = np.array(approx)>0.3
    
plt.scatter(e, fe, c=approx)

df = pd.DataFrame({'e':e, 'f(e)':fe})
```
